## Clustering

```python
#cluster on 2 features
df1=df[['Annual Income (k$)' , 'Spending Score (1-100)']]
df1.head()
```

```python
#scatter plot
sns.scatterplot(x = 'Annual Income (k$)',y = 'Spending Score (1-100)',  data = df1 ,s = 60 )
plt.title('Spending Score (1-100) vs Annual Income (k$)')
```

```python
from sklearn.cluster import KMeans
errors=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df1)
    errors.append(km.inertia_)
```

```python
#plot The results for elbow curve
plt.figure(figsize=(12,6))
plt.plot(range(1,11),errors)
plt.plot(range(1,11),errors, linewidth=2, color="red", marker ="8")
plt.xlabel(" NO. of clusters")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
#Taking 5 clusters
km1=KMeans(n_clusters=5)
#Fitting the input data
km1.fit(df1)
#predicting the labels of the input data
y=km1.predict(df1)
#adding the labels to a column named label
df1["label"] = y
#The new dataframe with the clustering done
df1.head()
```

```python
#Scatterplot of the clusters
sns.scatterplot(x = 'Annual Income (k$)',y = 'Spending Score (1-100)',data=df1,hue="label",s=60,palette=['#c76ca5','#30265c','brown','dodgerblue','purple'])
```

```python
#cluster on 3 features
df2=df[['Annual Income (k$)','Spending Score (1-100)','Age']]
df2.head()
```

```python
errors=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df2)
    errors.append(km.inertia_)
```

```python
#plot The results for elbow curve
plt.figure(figsize=(12,6))
plt.plot(range(1,11),errors)
plt.plot(range(1,11),errors, linewidth=2, color="red", marker ="8")
plt.xlabel(" NO. of clusters")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
#Taking 5 clusters
km1=KMeans(n_clusters=5)
#Fitting the input data
km1.fit(df2)
#predicting the labels of the input data
y=km1.predict(df2)
#adding the labels to a column named label
df2["label"] = y
#The new dataframe with the clustering done
df2.head()
```

```python
#3D Plot as we did the clustering on the basis of 3 input features
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df2.Age[df2.label == 0], df2["Annual Income (k$)"][df2.label == 0], df2["Spending Score (1-100)"][df2.label == 0], c='#c76ca5', s=60)
ax.scatter(df2.Age[df2.label == 1], df2["Annual Income (k$)"][df2.label == 1], df2["Spending Score (1-100)"][df2.label == 1], c='#30265c', s=60)
ax.scatter(df2.Age[df2.label == 2], df2["Annual Income (k$)"][df2.label == 2], df2["Spending Score (1-100)"][df2.label == 2], c='brown', s=60)
ax.scatter(df2.Age[df2.label == 3], df2["Annual Income (k$)"][df2.label == 3], df2["Spending Score (1-100)"][df2.label == 3], c='dodgerblue', s=60)
ax.scatter(df2.Age[df2.label == 4], df2["Annual Income (k$)"][df2.label == 4], df2["Spending Score (1-100)"][df2.label == 4], c='purple', s=60)
ax.view_init(30, 170)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
